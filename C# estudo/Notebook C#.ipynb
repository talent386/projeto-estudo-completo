## Hello World
```		
using System; // Necessário para usar Console.WriteLineusing System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;


namespace Helloworld // Agrupa o código sob o nome Helloworld
{
	class Program // Classe principal do programa
	{
		static void Main(string[] args) // Ponto de entrada
		{
			Console.WriteLine("Olá Mundo"); // Imprime mensagem no console
		}
	}
}

```
### Explicação do Código:
#### Importação de bibliotecas
```
using System;
using System.Collections.Generic;
using System.Linq;
using System.Text;
using System.Threading.Tasks;
```
Essas linhas importam **namespaces**, que são conjunstos de classes e funções, que fazem parte da biblioteca padrão do **.NET**

#### Namespace 
```
namespace Helloworld
```
`namespace ` é uma forma de organizar o código 
`Helloworld` é o nome escolhidso para o agrupamento do código

<sub>evita conflitos de nomes entre classes ou funções com o mesmo nome em diferentes partes do projeto</sub>

#### Classe
```
class Program
```
`class` define uma classe chamada `Progam`, todo código precisa estar dentro de uma classe e essa será a classe principal do programa.

#### Método Main 
```
static void Main(string[] args)
```

`static`: significa que o método pode ser chamado sem instanciar a classe `Program`

`void`: indica que o método não retorna nenhum valor

`string[] args`: um array de strings que pode receber argumentos da linha de comando.

#### Console WriteLine
```
Console.Writeline("Olá Mundo")
```
`Console` é uma classe da biblioteca System

`Writeline` escreve a linha no console e pula para a próxima linha.

#### Variação
```

namespace Helloworld 
{
	class Program
	{
		static void Main(string[] args)
		{
			System.Console.WriteLine("Olá Mundo!");
		}
	}
}
```

> Ao invés de chamar a classe no início do código, a classe é chamada apenas em pontos específicos. 

## Declaração de Variável
### Inicialização e Declaração na mesma linha:
```

    string textDataNascimento = Console.ReadLine();
    int dataNascimento = Convert.ToInt32(textDataNascimento);
```
### String para float

```
    string textMontante = Console.ReadLine();
    float textMontante = float.Parse(textMontante);
```
ou
```
    string textMontante = Console.ReadLine();
    float textMontante = Convert.ToSingle(textMontante);
```


### String para double 
```
    string textMontante = Console.ReadLine();
    float textMontante = Convert.ToDouble(textMontante);
```
## Operadores 

`var++` = +1

`var--` = -1 

`+` -> soma

`-` -> subtrai

`*` -> multiplica

`%` -> retorna o resto da divisão 

`/` -> divide

`=` -> atribui valor à variável

`+=` -> soma a variável por um valor

`-=` -> subtrai a variável por um valor

`*=	` -> multiplica a variável por um valor

`/=` -> divide a variável por um valor

`%=` -> retorna o resto da divisão da variável por um valor

`==`  -> igual a

`!=`  -> diferente de

`>`   -> maior que

`<`   -> menor que

`>=` -> maior ou igual a

`<=`  -> menor ou igual a

`&&`  -> E lógico (AND)

`||`  -> OU lógico (OR)

`!`   -> NÃO lógico (NOT)


## System.Text.Json
Biblioteca que permite **serializar** e **desserializar** objetos em JSON

> objeto -> JSON
> JSON -> objeto
